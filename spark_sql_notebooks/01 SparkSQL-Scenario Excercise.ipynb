```python
df = spark.read.format("csv").option("header","true").load("Files/Supermart Grocery Sales - Retail Analytics Dataset.csv")
# df now is a Spark DataFrame containing CSV data from "Files/salesdata/Supermart Grocery Sales - Retail Analytics Dataset.csv".
display(df)
```

```python
# Register the DataFrame as a temporary SQL table
df.createOrReplaceTempView("orders")
```

```python
%%sql
SELECT Order_ID, Customer_Name, Sales 
FROM orders;
```

**1. Write a query to find the total sales in each region.**

```python
%%sql
SELECT Region, SUM(sales) AS Total_Sales FROM orders 
GROUP BY Region
ORDER By Total_Sales ASC;
```

**2. Count how many orders were placed by each customer.**

```python
%%sql
select  customer_name, count(order_id) as total_orders from orders
group by customer_name
order by total_orders desc
```

**3. Find all orders with a discount of more than 20% but less than 30%**

```python
%%sql
select order_id,ROUND(SUM(discount * 100), 2) as discount from orders
where discount > 0.2 and discount < 0.3
GROUP by order_id
```

****

**4. List the top 3 orders by profit**

```python
%%sql
select * from orders
order by Profit desc
Limit 3
```

**5. Replace the category 'Snacks' with 'Quick Bites' in the result set**

```python
%%sql
SELECT category,
       CASE WHEN category = 'Snacks' THEN 'Quick bites' ELSE category END AS replaced_category
FROM orders;
```

```python
%%sql
SELECT category,
       REPLACE(category, 'Snacks', 'Quick bites') AS replaced_category
FROM orders;
```
